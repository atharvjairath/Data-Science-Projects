```python
def run(x_train,y_train):
    learning_rate = 0.1
    num_iterations = 1000
    y_pred = np.zeros(boston_data.shape[0])
    m = gd(x_train,y_train,learning_rate,num_iterations)
    for i in range(boston_data.shape[0]):
        y_pred[i] = predict(boston_data[i,:],m)
    #np.savetxt('project1.csv',y_pred,delimiter=",",fmt="%5f")
```

```python
def gd(x_train,y_train,learning_rate,num_iterations):
    x_train=scaler.fit_transform(x_train)
    q = x_train.shape[1]
    m = np.zeros(q)
    for i in range(num_iterations):
        m = step_gradient(x_train,y_train,learning_rate,m)
        print(i,cost(x_train,y_train,m))
        #print(m)
    return m
```

```python
def step_gradient(x_train,y_train,learning_rate,m):
    q = x_train.shape[1]
    M = len(x_train)
    m_slopes = np.zeros(len(m))
    for i in range(M):
        for j in range(q):
            m_slopes[j] += (-2/M)*(y_train[i] - ((m*x_train[i,:]).sum())) * x_train[i,j]
            #m[j] = m[j] -  learning_rate * (-2/M)*(y_train[i] - ((m*x_train[i,:]).sum())) * x_train[i,j]      
    m =  m - learning_rate * m_slopes
    return m
```

```python
def cost(x,y,m):
    total_cost = 0
    M = len(x)
    for i in range(M):
        total_cost += (1/M) * (y[i] - (m * x[i,:]).sum())**2
    return total_cost
```

```python
def predict(x,m):
    y_pred = (x.dot(m[:-1]) + m[-1])
    return y_pred
```
