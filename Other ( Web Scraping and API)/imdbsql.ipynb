```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
import sqlite3
c=sqlite3.connect("IMDB.sqlite")
data=pd.read_sql_query("Select * from earning",c)
data1=pd.read_sql_query("select * from IMDB",c)
data2=pd.merge(data,data1,on="Movie_id")
data2.Budget.replace("",0,inplace=True)
data2.Domestic.replace("",0,inplace=True)
data2.Worldwide.replace("",0,inplace=True)


def Yearcon(Title):
    return Title.split(" ")[-1].strip()


data2["Year"]=data2.Title.apply(Yearcon)


def removbrac(Year):
    return int(Year[-5:-1:1])
data2["Year"]=data2["Year"].apply(removbrac)

data2["profit1"]=data2["Domestic"]+data2["Worldwide"]
data3=data2.groupby('Year')["profit1"].max().index
data4=data2.groupby('Year')["profit1"].max().values

data5=[]
for i in range(len(data3)):
    s=data2[data2["profit1"]==data4[i]].Title.values[0]
    if "\xa0" in s:
        s=s.replace("\xa0"," ")
    data5.append(s)

for i in range(len(data4)):
    print(data3[i],data5[i])
```

```python
## Print the required output in given format
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
db=sqlite3.connect('IMDB.sqlite')
imdb=pd.read_sql_query("Select * from IMDB",db)
genre=pd.read_sql_query("Select * from genre",db)
df_merge = pd.merge(genre,imdb,on = "Movie_id")
df_merge["Budget"].replace('',0,inplace = True)
d = {}
for i in df_merge.index:
    if(df_merge['genre'][i] != ''):
        d[df_merge['genre'][i]] = d.get(df_merge['genre'][i],0) + df_merge["Budget"][i]

total_budget = sum(d.values())
for i in sorted(d):
    print(i,format(round(d[i]*100/total_budget , 2),'.2f'))
```
