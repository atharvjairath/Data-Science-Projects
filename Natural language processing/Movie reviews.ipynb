```python
from nltk.corpus import movie_reviews
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
movie_reviews.categories()
```

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[0:5]
```

```python
import random
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos=pos_tag([w])
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words            
```

```python
documents=[(clean_review(document),category) for document, category in documents]
```

```python
documents[0]
```

```python
training_documents=documents[0:1500]
testing_documents=documents[1500:]
```

```python
all_words=[]
for doc in training_documents:
    all_words+= doc[0]
```

```python
import nltk
freq=nltk.FreqDist(all_words)
common=freq.most_common(3000)
features=[ i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features={}
    word_set= set(words)
    for w in features:
        current_features[w]= w in word_set
    return current_features    
```

```python
training_data=[(get_feature_dict(doc),category) for doc, category in training_documents]
testing_data=[(get_feature_dict(doc),category) for doc, category in testing_documents]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classfier= NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classfier,testing_data)
```

```python
classfier.show_most_informative_features(15)
```
